```python
#find the numbers for each gene and then add the number to the file all_positive_final_version.csv
f1 = open("zebrafish_id_conversion_positive.txt","r")
transcripts = {} 
lines = f1.readlines()
#initialization
for line in lines:
    line = line.strip().split("\t")
    gene = line[2]
    transcripts[gene] = 0
#count numbers    
for line in lines:
    line = line.strip().split("\t")
    gene = line[2]
    transcripts[gene] = transcripts[gene]+1
#print len(transcripts.keys())
f1.close()


f2 = open("all_positive_final_version.csv","r")
f3 = open("all_positive_final_version_addTranscriptNumber.csv","w+")
for line in f2.readlines():
    line = line.strip().split(",")
    if line[2] in transcripts.keys():
        number = str(transcripts[line[2]])
        string = "\t".join([line[0],line[1],line[2],line[3],line[4],line[5],line[6],number])
        f3.write(string+"\n")
f2.close()
f3.close()
```

```python
#add point mutation position for each variant

import re
#p1 = re.compile("c")
p2 = re.compile("[A-Z]gg")
f = open("all_positive_final_version_addTranscriptNumber.csv","r")
f2 = open("all_positive_final_version_addPosition.csv","w+")
for line in f.readlines():
    line = line.strip().split(",")
    seq = line[7]
    first = re.search("c",seq).start()
    position = ''
    n20 = ''
    for m in p2.finditer(seq):
        pos = m.start()
        position = position+str(pos-first)+'/'
        n20 = n20+seq[pos-20:pos+3]+'|'
    line.append(position)
    line.append(n20)
    string =  "\t".join(line)
    #print string
    f2.write(string+"\n")
f.close()
#f2.close()
```
